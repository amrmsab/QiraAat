```python
for index, row in df.iterrows():
#     if index >= 3: break
    
#     f.write("<tr>\n")
    
#     x = [f"<td>{entry}</td>\n" for entry in list(row)]
#     f.write(''.join(x))
    
#     f.write("</tr>\n")
    
    f.write("<tr>\n")
    
    r = list(row)
    s = set(r[:-3])
    if '' in s:
        s = s - {''}
    s = list(s)
    c = np.arange(240,240-len(s)*30,-30)
    d = {list(s)[i]: c[i] for i in range(len(s))}
    
    j = 0
    colspan = 1
    while j < len(r):
        if j+1 < len(r) and r[j] == r[j+1]:
            pass
            colspan += 1
        else:
            if r[j] in d:
                f.write(f"<td style=\"background-color:rgb({d[r[j]]},{d[r[j]]},{d[r[j]]});\" colspan=\"{colspan}\">{r[j]}</td>\n")
            else:
                if j == 18: # verse column
                    f.write(f"<td style=\"text-align:right;\" colspan=\"{colspan}\">{r[j]}</td>\n")
                else:
                    f.write(f"<td colspan=\"{colspan}\">{r[j]}</td>\n")
            colspan = 1
        j += 1
    
    f.write("</tr>\n")
```
